```python
import random
import cv2
import matplotlib.pyplot as plt
```

```python
def add_noise(img):
    img = img.copy()

    # Getting the dimensions of the image
    row, col = img.shape

    # White noise (salt)
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):

        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)

        img[y_coord][x_coord] = 255

    # Black noise (pepper)
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):

        y_coord = random.randint(0, row - 1)
        x_coord = random.randint(0, col - 1)

        img[y_coord][x_coord] = 0

    return img


# Correct path
image_path = "D:/Desktop/Projects/OCR_project/images/johnnydepp.jpg"

img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Safety check
if img is None:
    print("Error: Image not found or path incorrect.")
else:
    noisy_img = add_noise(img)
    # Display images in notebook
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.title("Original Image")
    plt.imshow(img)
    plt.axis("off")
    cv2.imwrite("salt-and-pepper-image.jpg", noisy_img)
    plt.subplot(1,2,2)
    plt.title("Salt & Pepper Noise")
    plt.imshow(noisy_img, cmap='gray')
    plt.axis("off")

    plt.show()
```

Blurring the photos using three methods--
--gaussian
--median
--bilateral

```python
import cv2
from matplotlib import pyplot as plt

image_path = "D:/Desktop/Projects/OCR_project/Notebook/salt-and-pepper-image.jpg"


img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found.")
else:
    
    # Convert BGR → RGB for matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Gaussian Blur
    Gaussian = cv2.GaussianBlur(img, (15, 15), 0)
    Gaussian_rgb = cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.title("Salt and Pepper Image")
    plt.imshow(img_rgb)
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.title("Gaussian Blur Image")
    plt.imshow(Gaussian_rgb)
    plt.axis("off")

    plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image_path = "D:/Desktop/Projects/OCR_project/Notebook/salt-and-pepper-image.jpg"


img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found.")
else:
    
    # Convert BGR → RGB for matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Median Blur
    median = cv2.medianBlur(img, 11)  
    median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)  
    
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title("Salt and Pepper Image")
    plt.imshow(img_rgb)
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.title("Median Blur Image")
    plt.imshow(median_rgb)
    plt.axis("off")

    plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image_path = "D:/Desktop/Projects/OCR_project/Notebook/salt-and-pepper-image.jpg"


img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found.")
else:
    
    # Convert BGR → RGB for matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Bilateral Blur
    bilateral = cv2.bilateralFilter(img, 15, 150, 150)  
    bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB) 
    
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title("Salt and Pepper Image")
    plt.imshow(img_rgb)
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.title("Bilateral Blur Image")
    plt.imshow(bilateral_rgb)
    plt.axis("off")

    plt.show()
```
